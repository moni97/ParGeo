```python
import csv
import pandas as pd
```

```python
with open('output370.csv', mode ='r')as file:
   
  # reading the CSV file
    csvFile = csv.reader(file)
#     heading = ["Radius","Count","Calculated Entropy","Actual Entropy","Difference","Calculated Entropy Time", "Actual Entropy Time"];
    df = pd.DataFrame()
    # displaying the contents of the CSV file
    for lines in csvFile:
        line = lines[0].split(',')
        newRow = {"Radius": int(line[0]), "Count": int(line[1]), "Calculated Entropy":float(line[2]), "Actual Entropy":float(line[3]), "Entropy Difference": float(line[4]), "Time for calculated entropy":float(line[5]), "Time for actual entropy":float(line[6])}
        df = df.append(newRow,ignore_index=True)
```

```python
df
```

```python
radius = df['Radius'].unique()
```

```python
for rad in radius:
    chosenDf = df[df['Radius'] == rad]
    print("radius: ", rad, " entropy Difference: ", round(chosenDf['Entropy Difference'].mean(), 4), "time average: ", round(chosenDf["Time for calculated entropy"].mean(), 4))
```
